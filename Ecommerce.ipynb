```python
#Quantity ordered year wise
yearwisesales=df_ecommerce.groupby(df_ecommerce['InvoiceDate'].dt.year)['Quantity'].sum().reset_index().sort_values('Quantity', ascending=False)
yearwisesales
```

```python
data = yearwisesales['Quantity']
labels = yearwisesales['InvoiceDate']

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
